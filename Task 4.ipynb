## Step 3: Hypothesis Testing (Statistical Validation)

### 3.1: Define Hypothesis

H0 (Null Hypothesis): There is no significant difference in revenue between two months.
H1 (Alternative Hypothesis): There is a significant difference in revenue.

### 3.2: Perform Test in Python

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load dataset
df = pd.read_csv("online_retail_cleaned.csv")

# Convert date
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract months
df['Month'] = df['InvoiceDate'].dt.month

# Revenue
df['Revenue'] = df['Quantity'] * df['UnitPrice']

# Compare two months
jan = df[df['Month'] == 1]['Revenue']
feb = df[df['Month'] == 2]['Revenue']

# T-test
t_stat, p_value = ttest_ind(jan, feb)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

### 3.3: Interpretation

```python
-> If p-value < 0.05 → Significant difference
-> If p-value > 0.05 → Not significant
```
